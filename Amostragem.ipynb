```python
def amostragem_sistematica(dataset, amostra):
    intervalo = len(dataset) // amostra
    inicio = rd.randint(0, intervalo)
    indices = np.arange(7, len(dataset), step = 13)
    amostra_sistematica = dataset.iloc[indices]
    return amostra_sistematica
```

```python
amostra_sistematica = amostragem_sistematica(dataset, 400)
amostra_sistematica.head(10)
```
